## 클래스의 특징

### 1. 캡슐화
- 데이터와 메서드를 하나의 단위로 묶고, 외부에서 접근할 수 없도록 숨기는 개념.
- 접근 제한자
    - `public`: 어디서나 접근 가능 (기본).
    - `private` : 클래스 내부에서만 접근 가능 (앞에 __ 사용).

### 2. 상속
- 기존 클래스의 속성과 메서드를 물려받아 새로운 클래스를 만드는 방법.
- 코드의 재사용성을 높임.

### 3. 다형성
- 동일한 이름의 메서드가 여러 클래스에서 다르게 동작.
- 인터페이스의 일관성을 유지하면서 유연성을 제공합니다.

# 문제2

행복과 돈의 상관관계를 예측하는 모델의 class를 작성해보세요.

````python
class HappinessMoneyModel:
    def __init__(self, gradient=0.5, bias=2):
        """
        선형 회귀 모델 초기화
        :param gradient: 기울기 (default: 0.5)
        :param bias: 편향 (default: 2)
        """
        self.gradient = gradient
        self.bias = bias

    def predict(self, money):
        """
        돈에 기반하여 행복을 예측
        :param money: 돈의 양
        :return: 예측된 행복 값
        """
        return self.gradient * money + self.bias
````

````python

# 모델 인스턴스 생성 (기울기와 편향을 고정)
model = HappinessMoneyModel(gradient=0.8, bias=1.5)

# 데이터 포인트 생성 (돈, 행복)
data = [
    (10, model.predict(10)),
    (20, model.predict(20))
]

# 데이터 분리
money_values = [point[0] for point in data]
happiness_values = [point[1] for point in data]

````

### 문제3 : 학생들의 시험 점수 데이터를 클러스터링하여 유사한 성향의 그룹으로 분류하기

- 학생들의 수학 점수와 영어 점수를 기반으로 K-Means 클러스터링 모델을 구현하여 학생들을 유사한 성향의 그룹으로 분류합니다.
- 클러스터링 결과를 시각화하여 각 그룹을 색상으로 구분하고, 클러스터 중심을 표시합니다.

````python
import numpy as np

# 1. 데이터 생성
class Student:
    def __init__(self, name, math_score, english_score):
        self.name = name
        self.math_score = math_score
        self.english_score = english_score
        self.cluster = None  # 클러스터 번호를 저장할 속성

# 샘플 학생 데이터 생성
students = [
    Student("홍길동", 85, 78),
    Student("김철수", 92, 88),
    Student("이영희", 70, 65),
    Student("박민수", 60, 75),
    Student("최지우", 95, 98),
    Student("장준영", 55, 60),
    Student("오하늘", 80, 85),
    Student("유재석", 78, 82),
    Student("강호동", 88, 90),
    Student("송지효", 65, 70)
]

# 2. K-Means 클러스터링 모델 클래스 정의
class KMeansClustering:
    def __init__(self, n_clusters=2, max_iterations=100):
        self.n_clusters = n_clusters
        self.max_iterations = max_iterations
        self.centroids = []

    def initialize_centroids(self, data):
        # 데이터를 무작위로 선택하여 초기 클러스터 중심을 설정
        np.random.seed(42)  # 재현성을 위해 시드 고정
        random_indices = np.random.choice(len(data), self.n_clusters, replace=False)
        self.centroids = [data[idx] for idx in random_indices]

    def assign_clusters(self, data):
        for student in data:
            distances = [self.euclidean_distance(student, centroid) for centroid in self.centroids]
            student.cluster = distances.index(min(distances))

    def update_centroids(self, data):
        for i in range(self.n_clusters):
            cluster_members = [student for student in data if student.cluster == i]
            if cluster_members:
                avg_math = sum([student.math_score for student in cluster_members]) / len(cluster_members)
                avg_english = sum([student.english_score for student in cluster_members]) / len(cluster_members)
                self.centroids[i] = Student(f"Centroid {i+1}", avg_math, avg_english)

    def euclidean_distance(self, student1, student2):
        return np.sqrt((student1.math_score - student2.math_score)**2 + (student1.english_score - student2.english_score)**2)

    def fit(self, data):
        self.initialize_centroids(data)
        for _ in range(self.max_iterations):
            old_centroids = [ (c.math_score, c.english_score) for c in self.centroids ]
            self.assign_clusters(data)
            self.update_centroids(data)
            new_centroids = [ (c.math_score, c.english_score) for c in self.centroids ]
            if old_centroids == new_centroids:
                break

# 3. 모델 인스턴스 생성 및 학습
kmeans = KMeansClustering(n_clusters=2, max_iterations=100)
kmeans.fit(students)

# 4. 시각화
def plot_clusters(data, centroids, n_clusters):
    colors = ['skyblue', 'lightgreen', 'salmon', 'plum', 'gold']

    plt.figure(figsize=(10, 6))

    for i in range(n_clusters):
        cluster_members = [student for student in data if student.cluster == i]
        plt.scatter(
            [student.math_score for student in cluster_members],
            [student.english_score for student in cluster_members],
            color=colors[i % len(colors)],
            s=200,
            label=f'클러스터 {i+1}',
            edgecolors='k'
        )

    # 클러스터 중심 그리기
    for idx, centroid in enumerate(centroids):
        plt.scatter(
            centroid.math_score,
            centroid.english_score,
            color='red',
            marker='X',
            s=300,
            label=f'클러스터 중심 {idx+1}'
        )

    plt.title('학생들의 수학 점수와 영어 점수 기반 클러스터링', fontsize=18)
    plt.xlabel('수학 점수', fontsize=14)
    plt.ylabel('영어 점수', fontsize=14)
    plt.legend(fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

plot_clusters(students, kmeans.centroids, kmeans.n_clusters)
````
