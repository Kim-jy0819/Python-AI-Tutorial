# for문과 while문은 각각 언제 사용될까요?

for문은 반복 실행횟수를 명확히 알 때 사용하기 좋습니다. 예를 들면, 0부터 5까지 차례대로 더하라는 연산을 할 때는 for문이 편하니다.

반면에 while문은 반복 실행횟수가 명확하지 않을 때 사용합니다. 예를 들면 "사용자가 esc 키를 누르기 전까지 게임을 실행합니다."와 같은 프로그램을 실행할 때 사용됩니다.

# 반복의 제어 - break, continue

````python
for i in range(10):
  if i == 5:
    break # i가 5가 되면 반복 종료
  print (i)
print ('EOF') # 반복 종료 후 “EOP” 출력
````

````python
for i in range(10):
  if i == 5:
    continue # i가 5가 되면 i를 출력하지 않음
  print (i)
print ('EOF') # 반복 종료 후 “EOP” 출력
````

# 실전문제 5번

### 특정 질문에 대답하는 챗봇 만들기 (Python)
- 기능 설명:
    - 특정 질문에 대한 답변을 제공하는 간단한 챗봇입니다.
    - 미리 정의된 질문과 답변 목록을 사용합니다.
    - 사용자가 챗봇에 질문을 입력하면 챗봇이 조건문을 통해 적절한 답변을 출력합니다.

<br>

예시 실행:
```
안녕하세요! 저는 질문 대답 챗봇입니다. 질문해 주세요!
> 당신의 이름은 무엇인가요?
저는 파이썬 챗봇입니다.

> 날씨가 어떤가요?
죄송합니다, 그 질문은 이해하지 못했어요.

> 종료
안녕히 가세요!
```

````python
print("안녕하세요! 저는 질문 대답 챗봇입니다. 질문해 주세요!")
print("종료를 원하시면 '종료'라고 입력하세요.")

while True:
    user_input = input("> ").strip().lower()

    # 종료 조건
    if user_input == "종료":
        print("안녕히 가세요!")
        break

    # 질문-답변 매칭
    if user_input == "당신의 이름은 무엇인가요?":
        print("저는 파이썬 챗봇입니다.")
    elif user_input == "어떤 일을 하시나요?":
        print("저는 질문에 대답하고 정보를 제공합니다.")
    elif user_input == "몇 살인가요?":
        print("저는 나이를 먹지 않는 영원한 코드입니다!")
    elif user_input == "좋아하는 음식은 무엇인가요?":
        print("저는 코드만 먹어요! 🖥️")
    else:
        print("죄송합니다, 그 질문은 이해하지 못했어요.")
````
